## 1.4 Fit to get nonlinear E, v, and strain/stress rate 

```python
def quadratic_fit(x, y):
    p  = np.polyfit(x, y, 2)
    return p

def line_slope(x, y):
    p  = np.polyfit(x, y, 1)
    return p
```

```python
# use groupby fun. to solve strain/stress rate， d('ea')/d('t'), and d('q')/d('t')
grouped = Hdf.groupby(['Test','Path'], sort = False)
Hdf_Test_Path = grouped.apply(lambda x: quadratic_fit(x['ea'], x['q'])).to_frame(name ='fit1') #df['Efit'] = 
Hdf_Test_Path[['M2', 'M1','M0']] = Hdf_Test_Path['fit1'].apply(pd.Series) 
Hdf_Test_Path = Hdf_Test_Path.merge(grouped.apply(lambda x: quadratic_fit(x['ea'], x['er'])).to_frame(name ='fit2'), left_index = True, right_index = True)
Hdf_Test_Path[['V2', 'V1','V0']] = Hdf_Test_Path['fit2'].apply(pd.Series)
Hdf_Test_Path = Hdf_Test_Path.merge(grouped.apply(lambda x: line_slope(x['t'], x['ea'])).to_frame(name = 'fit3'), left_index = True, right_index = True)
Hdf_Test_Path[['Strain rate', 'Intercept1']] = Hdf_Test_Path['fit3'].apply(pd.Series)
Hdf_Test_Path = Hdf_Test_Path.merge(grouped.apply(lambda x: line_slope(x['t'], x['q'])).to_frame(name = 'fit4'), left_index = True, right_index = True)
Hdf_Test_Path[['Stress rate', 'Intercept2']] = Hdf_Test_Path['fit4'].apply(pd.Series)
Hdf_Test_Path = Hdf_Test_Path.round({'Strain rate': 3, 'Stress rate': 2})

Hdf_Test_Path = Hdf_Test_Path.drop(columns = ['Intercept1','Intercept2'] ) #+ ['fit' + str(w) for w in range(1,5)]

```

## 1.5 Use fitresult to calculate E and v

```python
# Calculate Static Young's modulus and Poisson's ratios with fitresult
est = []
# grouped = df.groupby(['Test','Path']) 
for name, group in grouped:
    Efit = 2 *   Hdf_Test_Path.loc[name,'M2'] * group['ea'].to_numpy() + Hdf_Test_Path.loc[name,'M1'] # E = 2* M2 * ea + M1
    vfit = - 2 * Hdf_Test_Path.loc[name,'V2'] * group['ea'].to_numpy() - Hdf_Test_Path.loc[name,'V1'] # V = -2 V2 * ea - V1
    ind = group.index.to_numpy()
    temp = np.column_stack((ind, Efit,vfit))
    if len(est) == 0:
        est = temp
    else:
        est = np.append(est,temp, axis =0)

Hdf_est = pd.DataFrame(data = est, columns = ['ind','Efit','vfit'])
Hdf_est['ind'] = Hdf_est['ind'].astype(int)
Hdf_est = Hdf_est.set_index('ind')
Hdf_Evfit = Hdf.merge(Hdf_est, how = 'left',left_index=True, right_index=True)
Hdf_Evfit.drop(Hdf_Evfit[Hdf_Evfit['Path']==''].index, inplace = True)

Hdf_Evfit.head(5)
```

## 1.6 Calculate the strain / stress magnitude, energy loss, Esec, and irreversible axial strain

```python
def mag(x):
    return x.max() - x.min()
```

```python
def irreversible(x):
    return x.iloc[-1] - x.iloc[0]
```

```python
"""This study does not aim to calculate energy loss, because each stress path is hand picked, and the strain-stress curve 
may not coincide at the top"""
def EnergyLoss(ea,q):
    idxmx = q.argmax()
    Eld = np.trapz(q[0:idxmx+1], ea[0:idxmx+1])
    Eul = np.trapz(q[idxmx+1:], ea[idxmx+1:])
    return Eld - Eul
```

```python
Hdf_Test_Path = Hdf_Test_Path.merge(Hdf.groupby(['Test','Path']).
                                  apply(lambda x: mag(x[['ea','q']])), 
                                   how = 'left', left_index= True, right_index= True
                                  ).rename(columns = {'ea':'Strain magnitude','q':'Stress magnitude'})
# df_test_Path.merge(df.groupby(['test','Path']).apply(lambda x: mag(x['q'])).reset_index(), how = 'left', on = ['test','Path']).rename(columns = {0:'Stress magnitude'})
```

```python
grouped = Hdf.groupby(['Test','Path'], sort = False)
temp = grouped.apply(lambda x: (mag(x['q'])/ mag(x['ea']))).to_frame(name = 'Esec')
temp['Test'] = temp.index.get_level_values(0)
temp['Path'] = temp.index.get_level_values(1)
temp.index = temp.index.set_names(['Test','Path'])
Hdf_Test_Path = Hdf_Test_Path.merge(temp,how = 'left', left_index = True, right_index = True)
```

## 1.7 Validate the polyfit method (QC estimated stress and strain)

```python
# Estimate the q and er with the fitting parameters
est = []

grouped = Hdf.groupby(['Test','Path'], sort = False)

for name, group in grouped:
    
#     print(name)
# #     print(df_Test_Path.loc[name,'M2'])
    q_est  = np.polyval(Hdf_Test_Path.loc[name,'fit1'], group['ea']) # 
    er_est = np.polyval(Hdf_Test_Path.loc[name,'fit2'], group['ea']) #
    ea_t   = np.polyval(Hdf_Test_Path.loc[name,'fit3'], group['t']) #
    q_t    = np.polyval(Hdf_Test_Path.loc[name,'fit4'], group['t']) #

    ind = group.index.to_numpy()
    temp = np.column_stack((ind, q_est, er_est, ea_t, q_t))
    if len(est) == 0:
        est = temp
    else:
        est = np.append(est,temp, axis =0)
```

## 2.4 Fit to get E, v, and strain/stress rate 

```python
# use groupby fun. to solve strain/stress rate， d('ea')/d('t'), and d('q')/d('t')
Vgrouped = Vdf.groupby(['Test','Path'], sort = False)
Vdf_Test_Path = Vgrouped.apply(lambda x: quadratic_fit(x['ea'], x['q'])).to_frame(name ='fit1') #Vdf['Efit'] = 
Vdf_Test_Path[['M2', 'M1','M0']] = Vdf_Test_Path['fit1'].apply(pd.Series) #Vdf_Test_Path[['M2', 'M1','M0','residuals']] = 
Vdf_Test_Path = Vdf_Test_Path.merge(Vgrouped.apply(lambda x: quadratic_fit(x['ea'], x['er'])).to_frame(name ='fit2'), left_index = True, right_index = True)
Vdf_Test_Path[['V2', 'V1','V0']] = Vdf_Test_Path['fit2'].apply(pd.Series)
Vdf_Test_Path = Vdf_Test_Path.merge(Vgrouped.apply(lambda x: line_slope(x['t'], x['ea'])).to_frame(name = 'fit3'), left_index = True, right_index = True)
Vdf_Test_Path[['Strain rate', 'Intercept1']] = Vdf_Test_Path['fit3'].apply(pd.Series)
Vdf_Test_Path = Vdf_Test_Path.merge(Vgrouped.apply(lambda x: line_slope(x['t'], x['q'])).to_frame(name = 'fit4'), left_index = True, right_index = True)
Vdf_Test_Path[['Stress rate', 'Intercept2']] = Vdf_Test_Path['fit4'].apply(pd.Series)
Vdf_Test_Path = Vdf_Test_Path.round({'Strain rate': 3, 'Stress rate': 2})

Vdf_Test_Path = Vdf_Test_Path.drop(['Intercept1','Intercept2'],axis = 1)
```

## 2.5 Use fitresult to calculate E and v

```python
"""Calculate Static Young's modulus and Poisson's ratios with fitresult"""
est = []
Vgrouped = Vdf.groupby(['Test','Path'], sort = False) 
for name, group in Vgrouped:
    Efit = 2 * Vdf_Test_Path.loc[name,'M2'] * group['ea'].to_numpy() + Vdf_Test_Path.loc[name,'M1']
    vfit = - 2 * Vdf_Test_Path.loc[name,'V2'] * group['ea'].to_numpy() - Vdf_Test_Path.loc[name,'V1']
    ind = group.index.to_numpy()
    temp = np.column_stack((ind, Efit,vfit))
    if len(est) == 0:
        est = temp
    else:
        est = np.append(est,temp, axis =0)

Vdf_est = pd.DataFrame(data = est, columns = ['ind','Efit','vfit'])
Vdf_est['ind'] = Vdf_est['ind'].astype(int)
Vdf_est = Vdf_est.set_index('ind')
Vdf_Evfit = Vdf.merge(Vdf_est, how = 'left',left_index=True, right_index=True)
Vdf_Evfit.drop(Vdf_Evfit[Vdf_Evfit['Path']==''].index, inplace = True)

Vdf_Evfit.head(5)
```

## 2.7 Validate the polyfit method (QC estimated stress and strain)

```python
"""Estimate the q and er with the fitting parameters"""
est = []
for name, group in Vgrouped:
#     print(name)
#     print(Vdf_Test_Path.loc[name,'M2'])
    q_est = np.polyval(Vdf_Test_Path.loc[name,'fit1'], group['ea']) # 
    er_est = np.polyval(Vdf_Test_Path.loc[name,'fit2'], group['ea']) #
    ea_t = np.polyval(Vdf_Test_Path.loc[name,'fit3'], group['t']) #
    q_t = np.polyval(Vdf_Test_Path.loc[name,'fit4'], group['t']) #

    ind = group.index.to_numpy()
    temp = np.column_stack((ind, q_est, er_est, ea_t, q_t))
    if len(est) == 0:
        est = temp
    else:
        est = np.append(est,temp, axis =0)

Vdf_est = pd.DataFrame(data = est, columns = ['ind','q_est','er_est','ea_t', 'q_t'])
Vdf_est['ind'] = Vdf_est['ind'].astype(int)
Vdf_est = Vdf_est.set_index('ind')
Vdf_withest = Vdf.merge(Vdf_est, how = 'left',left_index=True, right_index=True)
```

```python
Vgrouped = Vdf_withest.groupby(['Test','Path'])
R2 = [] # R-Squared
ind = Vdf_Test_Path.index
for name, group in Vgrouped:
    corr_mat = group.corr(method = 'pearson')
    R2_1, R2_2, R2_3, R2_4 = (corr_mat.loc['q','q_est']) ** 2, (corr_mat.loc['er','er_est']) ** 2, (corr_mat.loc['ea','ea_t']) ** 2, (corr_mat.loc['q','q_t']) **2
    temp = np.column_stack((R2_1, R2_2, R2_3, R2_4))
    if len(R2) == 0:
        R2 = temp
    else:
        R2 = np.append(R2,temp, axis =0)

Vdf_R2 = pd.DataFrame(data = R2, columns = ['R2_1', 'R2_2', 'R2_3', 'R2_4'], index = ind)
Vdf_Test_Path_R2 = Vdf_Test_Path.merge(Vdf_R2, how = 'left',left_index=True, right_index=True)
print(Vdf_Test_Path_R2.loc[(Vdf_Test_Path_R2.index.get_level_values(1) != ''),['R2_1', 'R2_2', 'R2_3', 'R2_4']].min())
print(Vdf_Test_Path_R2.loc[(Vdf_Test_Path_R2.index.get_level_values(1) != ''),['R2_1', 'R2_2', 'R2_3', 'R2_4']].idxmin())
```
